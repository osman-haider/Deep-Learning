```python
import numpy as np

# Step 1: Define dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input features
y = np.array([0, 0, 0, 1])
```

```python
y
```

```python
# Step 2: Initialize weights and bias
weights = np.random.rand(2)  # Random weights for 2 features
bias = np.random.rand()      # Random bias
learning_rate = 0.1          # Learning rate
```

```python
weights
```

```python
bias
```

```python
# Step 3: Train perceptron
for epoch in range(10):  # Train for 10 epochs
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        # Forward pass
        z = np.dot(X[i], weights) + bias  # Weighted sum
        y_pred = 1 if z >= 0 else 0       # Step function

        # Calculate error
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Error: {error}")
```
