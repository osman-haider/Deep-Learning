## Padding and Strides

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D

# Example Input
input_data = tf.random.normal([1, 6, 6, 1])  # Batch size 1, 6x6 grayscale image

# Model with Valid Padding
model_valid = Sequential([
    Conv2D(filters=1, kernel_size=(3, 3), strides=(1, 1), padding='valid')
])
output_valid = model_valid(input_data)

# Model with Same Padding
model_same = Sequential([
    Conv2D(filters=1, kernel_size=(3, 3), strides=(1, 1), padding='same')
])
output_same = model_same(input_data)

print(f"Input Shape: {input_data.shape}")
print(f"Output Shape (Valid Padding): {output_valid.shape}")
print(f"Output Shape (Same Padding): {output_same.shape}")
```
